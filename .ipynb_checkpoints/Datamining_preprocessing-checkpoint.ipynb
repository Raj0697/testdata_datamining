```python
import numpy as np
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
preprocesseddata.head(25)
```

```python
plt.scatter(preprocesseddata['home_score'],preprocesseddata['away_score'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(preprocesseddata[['home_score','away_score']])
y_predicted
```

```python
preprocesseddata['cluster'] = y_predicted
preprocesseddata.head()
```

```python
ppdata1 = preprocesseddata[preprocesseddata.cluster==0]
ppdata2 = preprocesseddata[preprocesseddata.cluster==1]
ppdata3 = preprocesseddata[preprocesseddata.cluster==2]

plt.scatter(ppdata1.home_score,ppdata1['away_score'],color='green')
plt.scatter(ppdata2.home_score,ppdata2['away_score'],color='red')
plt.scatter(ppdata3.home_score,ppdata3['away_score'],color='black')

plt.xlabel('homescore')
plt.ylabel('awayscore')
plt.legend(['homescore','homescore'])
```
